```python
from langchain.chat_models import ChatOpenAI
from langchain.callbacks import StreamingStdOutCallbackHandler
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory(return_messages=True)

# save_context(inputs, outputs) 메서드는 대화에서 사용자의 입력과 AI의 출력을 메모리에 저장하는 역할을 합니다.
# #여기서 inputs는 사용자 쪽 입력 데이터 딕셔너리이고, outputs는 AI의 응답 내용을 담은 딕셔너리입니다.
# #예를 들어, memory.save_context({"input": "Hello"}, {"output": "How are you?"})는 "Hello"라는 입력과 "How are you?"라는 출력 대화 쌍을 메모리에 저장합니다.
memory.save_context({"input": "Hello"}, {"output": "How are you?"})

# load_memory_variables는 저장된 대화 내용을 불러와 모델에게 그 맥락을 전달하는 용도로 사용됩니다.
# #이는 대화의 흐름을 유지하고 AI가 좀 더 일관된 응답을 할 수 있게 돕는 기능입니다.
memory.load_memory_variables({})
```
