```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    # model_name="gpt-5-nano",
    model_name="gpt-3.5-turbo",
    temperature=0.1,
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler(),
    ],
)

poem_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            (
                "너는 프로그램밍을 하는 개발자 이다. "
                "그리고 부업으로 시를 쓰는 시인이다."
                "좋아하는 시인은 하상욱, 김소연, 나희덕,백아온,윤동주 등이 있다"
                "그중 하상욱의 서울시 시집을 좋아한다."
            ),
        ),
        (
            "human",
            (
                "프로그래밍 언어인 {programLanguage}에 대한 시를"
                " 써주세요(3연으로 구성, 멋진 제목도 있어야함)."
            ),
        ),
    ]
)
poem_chain = poem_prompt | chat

desc_chef_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            (
                "당신은 문학 평론가 입니다."
                "주로 시를 평론하고, 고등학교에서 국어로 시를 가르칩니다."
                "학생들에게 수능 공부를 위해 시를 분석 하고 설명합니다."
            ),
        ),
        (
            "human",
            (
                "아래는 내가 쓴 시인데, 쉽게 알아 들을 정도로 설명하고 분석 해줘"
                "[시 제목] , 줄바꿈 2번 하고  설명헤줘"
                "{poem} "
            ),
        ),
    ]
)

desc_chain = desc_chef_prompt | chat

final_chain = {"poem": poem_chain} | desc_chain

final_chain.invoke({"programLanguage": "python"})
```
