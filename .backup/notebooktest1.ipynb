```python
from langchain.schema import BaseOutputParser

class CommaOutputParser(BaseOutputParser):
    def parse(self, text: str) -> list:
        item= text.strip().split(",")
        return list(map(str.strip, item ))
    
p = CommaOutputParser()
p.parse("apple, banana, cherry")
#['apple', ' banana', ' cherry']
```

```python
template = ChatPromptTemplate.from_messages([
    ("system","You are a list generating machine. Everything you are asked will be answered with a comma separated list of max {max_items} in lowercase. Do not relpy with anything else."),
    ("human","{question}"),
])

prompt = template.format_messages(question="What are the top programming languages in 2024?",max_items=5)

result = chat.predict_messages(prompt)#AIMessage(content='python, javascript, java, c#, typescript')

p= CommaOutputParser()
p.parse(result.content)
#['python', 'javascript', 'java', 'c#', 'typescript']

```

```python
#p= CommaOutputParser()
chain = template | chat | p

chain.invoke({
    "question":"What are the top programming languages in 2000?",
    "max_items":5
})  #input type = Dictionary
#['java', 'c', 'c++', 'visual basic', 'perl']   

```

```python
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    model_name="gpt-5-nano",
    temperature=1,
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler(),
    ],
)

chef_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a world-class international chef. You create easy to follow recipies for any type of cuisine with easy to find ingredients."),
        ("human", "I want to cook {cuisine} food.")
    ]
)

chef_chain = chef_prompt | chat

veg_chef_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a vegetarian chef specialized on making traditional recipies vegetarian. You find alternative ingredients and explain their preparation. You don't radically modify the recipe. If there is no alternative for a food just say you don't know how to replace it.",
        ),
        ("human", "{recipe}, please reply in korean."),
    ]
)

veg_chain = veg_chef_prompt | chat

final_chain = {"recipe": chef_chain} | veg_chain

final_chain.invoke({"cuisine":"korean"})
```

```python
from langchain.prompts.few_shot import FewShotPromptTemplate


examples = [
    {
        "question": "What do you know about France?",
        "answer": (
            """
    Here is what I know:
    Capital: Paris
    Language: French
    Food: Wine and Cheese
    Currency: Euro
    """
        ),
    },
    {
        "question": "What do you know about Italy?",
        "answer": (
            """
    I know this:
    Capital: Rome
    Language: Italian
    Food: Pizza and Pasta
    Currency: Euro
    """
        ),
    },
    {
        "question": "What do you know about Greece?",
        "answer": (
            """
    I know this:
    Capital: Athens
    Language: Greek
    Food: Souvlaki and Feta Cheese
    Currency: Euro
    """
        ),
    },
]

examples_template = """
    Human:{question},
    AI:{answer}
""" 

"""
t = PromptTemplate.from_template("What is the capital of {country}")
둘은 같다.
t = PromptTemplate(
    template="What is the capital of {country}",
    input_variables=["country"],
)

"""
examples_prompt = PromptTemplate.from_template(examples_template)
#둘은 같다.
examples_prompt = PromptTemplate(
    template=examples_template,
    input_variables=["question","answer"],
)

#가져온 [예제]를 prompt 를 사용해서 형식화
prompt = FewShotPromptTemplate(
    example_prompt=examples_prompt,
    examples=examples,
    #사용자 질문이 어떻게 보여줄지 지정...
    suffix="Human: What do you know about {country}? please reply in korean.\nAI:" ,
    #어떤 변수를 suffix에 넣을지 지정
    input_variables=["country"],    
)


#prompt.format(country="Spain")
#'\n    Human:What do you know about France?,\n    AI:\n    Here is what I know:\n    Capital: Paris\n    Language: French\n    Food: Wine and Cheese\n    Currency: Euro\n    
# \n\n\n\n    Human:What do you know about Italy?,\n    AI:\n    I know this:\n    Capital: Rome\n    Language: Italian\n    Food: Pizza and Pasta\n    Currency: Euro\n    \n\n\n\n    Human:What do you know about Greece?,\n    AI:\n    I know this:\n    Capital: Athens\n    Language: Greek\n    Food: Souvlaki and Feta Cheese\n    Currency: Euro\n    
# \n\n\nHuman: What do you know about Spain?\nAI:'

chain = prompt | chat
chain.invoke({"country":"Germany"})
```

```python
from langchain.prompts.few_shot import FewShotChatMessagePromptTemplate
from langchain.prompts import ChatPromptTemplate


examples = [
    {
        "country": "France",
        "answer": (
            """
        Here is what I know:
        Capital: Paris
        Language: French
        Food: Wine and Cheese
        Currency: Euro
        """
        ),
    },
    {
        "country": "Italy",
        "answer": (
            """
        I know this:
        Capital: Rome
        Language: Italian
        Food: Pizza and Pasta
        Currency: Euro
        """
        ),
    },
    {
        "country": "Greece",
        "answer": (
            """
        I know this:
        Capital: Athens
        Language: Greek
        Food: Souvlaki and Feta Cheese
        Currency: Euro
        """
        ),
    },
]

examples_prompt = ChatPromptTemplate.from_messages([
    ("human", "What do you know about {country}?"),
    ("ai", "{answer}"),
])  

examples_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=examples_prompt,
    examples=examples,
    input_variables=["country"],
)
#a = examples_prompt.invoke({"country":"한국"})
final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a geography expert, you give short answers. 하지만 추가로 전문가적으로 꼭 알아야할 내용이 있으면 알려줘"),
        examples_prompt,
        ("human", "What do you know about {country}? please reply in korean."),
    ]
)

chain = final_prompt | chat
#chain = examples_prompt | chat
chain.invoke({"country":"korea"})
```
