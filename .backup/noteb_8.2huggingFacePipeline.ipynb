```python
from langchain.llms.huggingface_pipeline import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('gpt2')
#tokenizer = AutoTokenizer.from_pretrained('tiiuae/falcon-7b-instruct')
tokenizer.pad_token = tokenizer.eos_token
system_message = "You are an assistant that translates from English to Korean.\n"
prompt = PromptTemplate.from_template(system_message + "Please reply in Korean. Define this word : {word} ")

lim = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    pipeline_kwargs={"max_new_tokens": 150}, 
)

chain = prompt | lim

chain.invoke({"word": "school"})
```
